```julia
using Roots
its = zeros(length(ps),length(Ls))
err = zeros(length(ps),length(Ls))
x0 = Array(-1:0.001:1)
sols = zeros(length(ps),length(x0))
leg = []
u0 = 0
u1 = 0
M = 0
for j=1:length(ps)
    p = ps[j]
    fname = @sprintf("generated/SOL1d-%d-%d.jld2",p*10,Ls[end])
    SOL0=load(fname)["SOL"]
    M = SOL0.B.M
    u0 = M.D[1]*SOL0.x
    if p==1
        u0 = -ones(Float64,size(u0))
        u0[end] = 1.0
    else
        q = 1/(1-1/p)
        a = (0.5/(p^p*(p-1)^(1-p)))^(1/(p-1))
        f9 = b->a*((1+b)^q-(b-1)^q)-2
        b = fzero(f9,2.0)
        c = -1-a*(b-1)^q
        u0 = a.*((M.x[:,1] .+ b).^q) .+ c
    end
    sols[j,:] = interp1d(M,u0,x0)
    push!(leg,@sprintf("p=%.1f",p))
    for k=1:length(Ls)
        L = Ls[k]
        fname = @sprintf("generated/SOL1d-%d-%d.jld2",p*10,L)
        SOL=load(fname)["SOL"]
        u1 = interp1d(SOL.B.M,SOL.B.M.D[1]*SOL.x,M.x)
        err[j,k] = sqrt(dot(M.w[:,end],(u1-u0).^2))
        its[j,k] = sum(SOL.its)
    end
end
```

```julia
semilogy(Ls,its')
xlabel(L"$L$")
ylabel("Newton iterations")
title("Newton iterations in 1d", fontsize=16)
legend(leg)
savefig("generated/iter1d.eps")
```

```julia
semilogy(Ls[1:end],err[:,1:end]')
xlabel(L"$L$")
ylabel(L"$\|u_L^*-u^*\|_{L^2}$")
title("Error in 1d",fontsize=16)
#legend(leg)
savefig("generated/error1d.eps")
```

```julia
plot(x0,sols')
xlabel(L"$x$")
title(L"Solutions in 1d with $L=80$",fontsize=16)
legend(leg)
savefig("generated/sol1d.eps")
```
